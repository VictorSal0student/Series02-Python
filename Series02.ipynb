# Series02-Python : Sinusoidal Waves

The equation for a sinusoidal wave is given by:

$$ y(t) = A \sin(2 \pi f t + \phi) $$

Where:
- \( y(t) \) is the displacement at time \( t \)
- \( A \) is the amplitude (the peak value) of the wave
- \( f \) is the frequency (the number of cycles per unit time)
- \( \phi \) is the phase shift (the horizontal shift of the wave)
- \( t \) is the time variable

```python
# implement the sinusoidal wave equation in Python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
A = 1.0            # Amplitude
T = 1.0            # Period in seconds
u = 2 * np.pi / T  # Angular frequency
phi = 0            # Phase shift

# Time array
n_periods = 1                       # Number of periods to display
Duration = n_periods * T            # Duration in seconds
t = np.linspace(0, Duration, 1000)  # Time from 0 to Duration

# Sinusoidal wave equation
y = A * np.sin(u * t + phi)

# Plotting the wave
plt.figure(figsize=(6, 4))
plt.plot(t, y, label='y(t) = A sin(ωt + φ)')
plt.title('Sinusoidal Wave')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (y)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```
